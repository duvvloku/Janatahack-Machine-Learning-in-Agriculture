```python
df["Number_Weeks_Used"].fillna(df["Number_Weeks_Used"].mean(), inplace=True)
```

```python
X_train = df.drop(["ID","Crop_Damage"] , axis=1)
y_train = df.pop("Crop_Damage")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbcl = GradientBoostingClassifier(n_estimators = 500,learning_rate=0.2,random_state=1)
gbcl = gbcl.fit(X_train, y_train)
```

```python
print(gbcl.score(X_train, y_train))
```
